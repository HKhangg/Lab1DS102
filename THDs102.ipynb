```python
def convert_month(month):
    month_dict = {
        "jan": 1,
        "feb": 2,
        "mar": 3,
        "apr": 4,
        "may": 5,
        "jun": 6,
        "jul": 7,
        "aug": 8,
        "sep": 9,
        "oct": 10,
        "nov": 11,
        "dec": 12
    }
    return month_dict[month]
```

```python
def convert_day(day):
    day_dict = {
        "mon": 0,
        "tue": 1,
        "wed": 2,
        "thu": 3,
        "fri": 4,
        "sat": 5,
        "sun": 6
    }
    return day_dict[day]
```

```python
df["month"] = df["month"].apply(convert_month)
```

```python
df["day"] = df["day"].apply(convert_day)
```

```python
df.head(10)
```

```python
class LinearRegression:
    def fit(self, X, y):
        cov = X.T @ X
        inv_cov = np.linalg.inv(cov)
        self.theta = inv_cov @ (X.T @ y)
    def predict(self, X):
        y_predict = self.theta.T @ X.T
        return y_predict
    def rmse(self,y_pred, y_test):
        rmse = np.square(y_pred - y_test).mean()**0.5
        return rmse
```

```python
data = df.to_numpy()
```

```python
np.random.shuffle(data)
```

```python
train_size = int(0.8*len(df))
train_size
```

```python
X_y_train = data[:train_size]
```

```python
X_y_test = data[train_size:]
```

```python
X_train = X_y_train[:,:12]
y_train = X_y_train[:,-1]
```

```python
X_test = X_y_test[:,:12]
y_test = X_y_test[:,-1]
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
y_predict = lr.predict(X_test)
lr.rmse(y_predict,y_test)
```
