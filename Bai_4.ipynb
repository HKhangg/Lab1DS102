```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression 
```

```python
df = pd.read_csv("D:\\Learn\\DS102ML\\forest+fires\\processing_forestfires.csv")
df.head(10)
```

```python
data = df.to_numpy()
```

```python
train_size = int(0.8*len(df))
train_size
```

```python
X_y_train = data[:train_size]
X_y_test = data[train_size:]
```

```python
X_train = X_y_train[:,:12]
y_train = X_y_train[:,-1]
```

```python
X_test = X_y_test[:,:12]
y_test = X_y_test[:,-1]
```

```python
regr = LinearRegression() 
  
regr.fit(X_train, y_train)
```

```python
y_predicted = regr.predict(X_test)
```

```python
def rmse(y_pred, y_test):
        rmse = np.square(y_pred - y_test).mean()**0.5
        return rmse
```

```python
print("RMSE:", rmse(y_predicted, y_test))
```
